#### Finding out how much amount is cancelled and uncancelled

Getting the percentage to which cancellation is occurring

```python
# depending on hotels lets come to know which one has more cancelleation rates
```

```python
plt.figure(figsize=(8,4))
# ax1=sns.countplot(x='hotel',hue='is_canceled',data=df,palette='Blues')
# legend_labels,_=ax1.get_legend_handles_labels()
# ax1.legend(bbox_to_anchor(1,1))
# plt.title('Reservation status in different hotels',size=20)
# plt.xlabel('hotel')
# plt.ylabel('number of reservations')

plt.figure(figsize=(8,4))
ax1 = sns.countplot(x='hotel', hue='is_canceled', data=df, palette='Blues')

# Get the labels for the legend
legend_labels, _ = ax1.get_legend_handles_labels()

# Fix the bbox_to_anchor reference
ax1.legend(bbox_to_anchor=(1, 1))  # Correct syntax here

# Set title and axis labels
plt.title('Reservation status in different hotels', size=20)
plt.xlabel('hotel')
plt.ylabel('number of reservations')
plt.legend(['Not Cancelled','Cancelled'])

# Display the plot
plt.show()
```

```python
# light blue not cancelled
#  dark blue cancelled
# we can come to the conclusion that city hotrl is experiencing more 
# cancellations as well as more bookingd 
```

```python
resort_hotel=df[df['hotel']=='Resort Hotel']
resort_hotel['is_canceled'].value_counts(normalize=True)
```

```python
city_hotel=df[df['hotel']=='City Hotel']
city_hotel['is_canceled'].value_counts(normalize=True)
```

```python
resort_hotel=resort_hotel.groupby('reservation_status_date')[['adr']].mean()
city_hotel=city_hotel.groupby('reservation_status_date')[['adr']].mean()
```

```python
plt.figure(figsize=(20,8))
plt.title("Average Daily rate in City and resort Hotel",fontsize=30)
plt.plot(resort_hotel.index,resort_hotel['adr'],label="Resort Hotel")
# plt.legend(fontsize=20)
plt.plot(city_hotel.index,city_hotel['adr'],label="City Hotel")
plt.legend(fontsize=20)
plt.show()
         
```

```python
# find out month wise reservation and canceeling
```

```python
df['month']=df['reservation_status_date'].dt.month
plt.figure(figsize=(16,8))
ax1=sns.countplot(x='month',hue='is_canceled',data=df,palette='bright')
legend_labels, _ = ax1.get_legend_handles_labels()

# Fix the bbox_to_anchor reference
ax1.legend(bbox_to_anchor=(1, 1))  # Correct syntax here

# Set title and axis labels
plt.title('Reservation status per month', size=20)
plt.xlabel('month')
plt.ylabel('number of reservations')
plt.legend(['Not Cancelled','Cancelled'])
```

```python
# from this we come to know more number of cancellations occurred in August and 
#  more bookings in month of January
```

```python
# we can conclude that maybe prices were high in january which is why
#  customers may have given it  a second thought to cancel it out
# it is also possible that due to less price in August ,cancellation were less too
```

```python
# So to confirm this we are plotting Average Daily rate
plt.figure(figsize=(15,8))
plt.title('ADR per month',fontsize=30)
# sns.barplot('month','adr',data=df[df['is_canceled']==1].groupby('month')[['adr']].sum().reset_index())
sns.barplot(x='month', y='adr', data=df[df['is_canceled'] == 1].groupby('month')[['adr']].sum().reset_index())

plt.show()
```

```python
# thus it is now confirmed that 1 january had more cancellations and it was because of high price rates 
# wheras august had less cancellation due to less average daily rate
# what insight we could extract from data is tthere is a correlation present in 
# between average daily rate i.e. prices and cancellations
```

## Cancellation on the basis of countires (top 10)

```python
cancelled_data=df[df['is_canceled']==1]
top10=cancelled_data['country'].value_counts()[:10]
plt.figure(figsize=(8,8))
plt.title('Top 10 countries with reservation canceled')
plt.pie(top10,autopct='%.2f',labels=top10.index)
plt.show()
```

```python
# Thus we can draw conclusion that PORTUGAL is performing more cancellations
#  thus it can be advised to upgrade facilities, provide promotional discounts ,run campaigns etc
```

```python
cancelled_df_adr = cancelled_data.groupby('reservation_status_date')[['adr']].mean()  
cancelled_df_adr.reset_index(inplace = True)  
cancelled_df_adr.sort_values('reservation_status_date', inplace = True)  

not_cancelled_data = df[df['is_canceled'] == 0]  
not_cancelled_df_adr = not_cancelled_data.groupby('reservation_status_date')[['adr']].mean()  
not_cancelled_df_adr.reset_index(inplace = True)  
not_cancelled_df_adr.sort_values('reservation_status_date', inplace = True)  

plt.figure(figsize = (20,6))  
plt.title('Average Daily Rate')  
plt.plot(not_cancelled_df_adr['reservation_status_date'], not_cancelled_df_adr['adr'], label = 'not cancelled')  
plt.plot(cancelled_df_adr['reservation_status_date'], cancelled_df_adr['adr'], label = 'cancelled')  
plt.legend()
```

```python
cancelled_df_adr = cancelled_df_adr[(cancelled_df_adr['reservation_status_date']>'2016') & (cancelled_df_adr['reservation_status_date']<'2017-09')]  
not_cancelled_df_adr = not_cancelled_df_adr[(not_cancelled_df_adr['reservation_status_date']>'2016') & (not_cancelled_df_adr['reservation_status_date']<'2017-09')]
```
